## Text Summarization Using Prompt Template

```python
from langchain.chains import LLMChain
# LLMChain is a component in the Langchain library that allows you to create a chain of language model calls. 
# It is used to structure and manage the flow of information between different language model components, 
# enabling complex interactions and processing sequences. LLMChain can be used to build applications that 
# require multiple steps of language model processing, such as summarization, translation, or dialogue systems.

from langchain.prompts import PromptTemplate

generictemplate = """ Write a summary of the following speech
Speech: {speech}
Translate this precise summary to language: {language} """

prompt = PromptTemplate(
    input_variables = ['speech','language'],
    template = generictemplate
)
prompt
```

```python
# Formatting the PromptTemplate
complete_prompt = prompt.format(speech=speech, language = 'Hindi')
```

```python
llm.get_num_tokens(complete_prompt)
```

```python
from langchain_core.output_parsers import StrOutputParser

chain = prompt|llm|StrOutputParser()
summary = chain.invoke({'speech':speech,"language":"Hindi"})
```

```python
summary
```
